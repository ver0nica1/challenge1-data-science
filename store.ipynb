### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "base-de-datos-challenge1/tienda_1 .csv"
url2 = "base-de-datos-challenge1/tienda_2.csv"
url3 = "base-de-datos-challenge1/tienda_3.csv"
url4 = "base-de-datos-challenge1/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 1. Análisis de facturación

En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
ingresoTotal = tienda['Precio'].sum()
ingresoTotal
```

```python
ingresoTotal2 = tienda2['Precio'].sum()
ingresoTotal2
```

```python
ingresoTotal3 = tienda3['Precio'].sum()
ingresoTotal3
```

```python
ingresoTotal4 = tienda4['Precio'].sum()
ingresoTotal4
```

```python
print(f'El ingreso total de la tienda 1 es: ${ingresoTotal:,.2f} pesos')
print(f'El ingreso total de la tienda 2 es: ${ingresoTotal2:,.2f} pesos')
print(f'El ingreso total de la tienda 3 es: ${ingresoTotal3:,.2f} pesos')
print(f'El ingreso total de la tienda 4 es: ${ingresoTotal4:,.2f} pesos')
```

```python
tiendas = [ 'Tiendas 1', 'Tiendas 2', 'Tiendas 3', 'Tiendas 4' ]
ingresos = [ingresoTotal, ingresoTotal2, ingresoTotal3, ingresoTotal4]
colores = ['#ae115a', '#e91969', '#ff4f83', '#ff9dbd']

plt.figure(figsize=(14, 6))
barras = plt.bar(tiendas, ingresos, color=colores)
plt.subplot(1, 1, 1)
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos Totales (pesos)')

for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width() / 2, altura, f'${altura:,.2f}', ha='center', va='bottom')
```

# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
ventas_tienda1 = tienda.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print(ventas_tienda1)

new_var = plt.figure(figsize=(14, 6))
colores = ['#590D22', '#800F2F', '#A4133C', '#C9184A', '#FF4D6D', '#FF758F', '#FF8FA3', '#FFB3C1']
plt.pie(ventas_tienda1, labels=ventas_tienda1.index, autopct='%1.1f%%', colors=colores)
plt.subplot(1, 1, 1)
plt.title('Distribución de Ventas por Categoría en la Tienda 1')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
calificacion_tienda1 = tienda['Calificación'].mean()
calificacion_tienda2 = tienda2['Calificación'].mean()
calificacion_tienda3 = tienda3['Calificación'].mean()
calificacion_tienda4 = tienda4['Calificación'].mean()

print(f'La calificación promedio de la tienda 1 es: {calificacion_tienda1:.2f}')
print(f'La calificación promedio de la tienda 2 es: {calificacion_tienda2:.2f}')
print(f'La calificación promedio de la tienda 3 es: {calificacion_tienda3:.2f}')
print(f'La calificación promedio de la tienda 4 es: {calificacion_tienda4:.2f}')

calificaciones = [calificacion_tienda1, calificacion_tienda2, calificacion_tienda3, calificacion_tienda4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
colores = ['#ae115a', '#e91969', '#ff4f83', '#ff9dbd']

plt.figure(figsize=(14, 6))
barras = plt.bar(tiendas, calificaciones, color=colores)
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Calificación Promedio')

for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width() / 2, altura, f'{altura:.2f}', ha='center', va='bottom')

plt.show()
```

# 5. Envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
envio_promedio = tienda['Costo de envío'].mean()
envio_promedio2 = tienda2['Costo de envío'].mean()
envio_promedio3 = tienda3['Costo de envío'].mean()
envio_promedio4 = tienda4['Costo de envío'].mean()

print(f'El costo de envío promedio en todas la teinda 1 es: ${envio_promedio:.2f} pesos')
print(f'El costo de envío promedio en todas la teinda 2 es: ${envio_promedio2:.2f} pesos')
print(f'El costo de envío promedio en todas la teinda 3 es: ${envio_promedio3:.2f} pesos')
print(f'El costo de envío promedio en todas la teinda 4 es: ${envio_promedio4:.2f} pesos')
```

```python
plt.figure(figsize=(14, 6))
envios = [envio_promedio, envio_promedio2, envio_promedio3, envio_promedio4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
colores = ['#ae115a', '#e91969', '#ff4f83', '#ff9dbd']
barras = plt.barh(tiendas, envios, color=colores)
plt.title('Costo de Envío Promedio por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Costo de Envío Promedio (pesos)')

for barra in barras:
    ancho = barra.get_width()
    plt.text(ancho, barra.get_y() + barra.get_height() / 2, f'${ancho:.2f}', ha='left', va='center')
    
plt.show()
```

## Introducción

El objetivo de este análisis fue evaluar el desempeño de las cuatro tiendas para definir cuál debería vender el Sr. Juan. Para ello, se compararon cinco factores clave: ingresos totales, comportamiento de ventas por categoría, calificación promedio de clientes, productos más y menos vendidos, y costo de envío promedio.

## Desarrollo (hallazgos principales)

### 1) Ingresos totales
- **Tienda 1:** $1.150.880.400
- **Tienda 2:** $1.116.343.500
- **Tienda 3:** $1.098.019.600
- **Tienda 4:** $1.038.375.700

La **Tienda 4** presenta el menor ingreso total, con una diferencia importante frente a las demás.

### 2) Categorías más y menos vendidas
- En las cuatro tiendas, la categoría más vendida es **Muebles**.
- Las categorías menos vendidas se concentran en **Artículos para el hogar** (Tiendas 1, 2 y 3) e **Instrumentos musicales** (Tienda 4).

Esto muestra un patrón comercial similar entre tiendas, por lo que la diferenciación principal se observa más en ingresos y percepción del cliente.

### 3) Calificación promedio de clientes
- **Tienda 1:** 3,98
- **Tienda 2:** 4,04
- **Tienda 3:** 4,05
- **Tienda 4:** 4,00

La Tienda 3 obtiene la mejor valoración y la Tienda 1 la más baja. La Tienda 4 queda en un nivel intermedio.

### 4) Productos más y menos vendidos
Cada tienda tiene productos líderes distintos, pero en general no se observa una ventaja competitiva clara de la Tienda 4 frente a las demás. Además, sus productos menos vendidos indican oportunidades de mejora en rotación de inventario.

### 5) Costo de envío promedio
- **Tienda 1:** $26.018,61
- **Tienda 2:** $25.216,24
- **Tienda 3:** $24.805,68
- **Tienda 4:** $23.459,46

La Tienda 4 tiene el envío promedio más bajo (fortaleza logística/comercial). Sin embargo, este punto no compensa su menor facturación total.

## Conclusión y recomendación

Con base en el análisis integral, la recomendación es que **el Sr. Juan venda la Tienda 4**.

La justificación principal es su **menor ingreso total**, que es el indicador más crítico para la sostenibilidad del negocio. Aunque la Tienda 4 muestra un costo de envío favorable y una calificación aceptable, su desempeño global queda por debajo del resto. En contraste, las Tiendas 2 y 3 combinan mejores calificaciones y mayores ingresos, mientras que la Tienda 1, aun con menor calificación, mantiene la mayor facturación.

Por tanto, considerando fortalezas y debilidades de forma objetiva, la Tienda 4 es la opción más coherente para ser vendida.
